```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output1.csv')
```

```python
dynamic_means = df.loc[df['Type'] == 'Dynamic'].groupby('Size').mean().drop('Price', axis=1)
greedy_means = df.loc[df['Type'] == 'Greedy'].groupby('Size').mean().drop('Price', axis=1)
dynamic_sdv = df.loc[df['Type'] == 'Dynamic'].groupby('Size').std().drop('Price', axis=1)
greedy_sdv = df.loc[df['Type'] == 'Greedy'].groupby('Size').std().drop('Price', axis=1)
```

```python
sizes = [4,8,12,16,20,30,40,50,60,70,80,90,100,120,140,160,180,200]
```

```python
plt.figure(figsize=(10,5))
plt.plot(sizes, greedy_means, lw = 3, label='Média/Guloso')
plt.plot(sizes, dynamic_means, lw=3, color='red', label='Média/Prog. Dinâmica')
plt.title('Médias de Tempo de execução')
plt.xlabel('Número de Ilhas')
plt.ylabel('Tempo em microssegundos')
plt.legend()
plt.show()
plt.figure(figsize=(10,5))
plt.plot(sizes, greedy_sdv, lw = 3, label='Desvio/Guloso', color = 'purple')
plt.plot(sizes, dynamic_sdv, lw=3, color='orange', label='Desvio/Prog. Dinâmica')
plt.title('Desvios Padrões de Tempo de execução')
plt.xlabel('Número de Ilhas')
plt.ylabel('Tempo em microssegundos')
plt.legend()
plt.show()
```
